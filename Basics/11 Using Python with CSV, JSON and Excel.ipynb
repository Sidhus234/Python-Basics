```python
df4.head(10)
```

```python
df4["Geocode Address"] = df4["Address"] + ", " + df4["City"] + ", " + df4["State"] + ", " + df4["Country"]
df4
```

```python
df4["Coordinates"] = df4["Geocode Address"].apply(nom.geocode)
df4
```

```python
df4["Latitude"] = df4["Coordinates"].apply(lambda x: x.latitude)
df4["Longitude"] = df4["Coordinates"].apply(lambda x: x.longitude)
```
